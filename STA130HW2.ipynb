```python
#Q4: Watch the embedded video tutorial on Markov chains in the next Jupyter cell below to understand their application and relevance for ChatBots; then, after watching the video, start a new ChatBot session by prompting that you have code that creates a "Markovian ChatBot"; show it the first version of the "Markovian ChatBot code" below; and interact with the ChatBot session to make sure you understand how the original first version of the "Markovian ChatBot code" works




#Here are the codes:
import random
import re
from collections import defaultdict

class MarkovianChatBot:
    def __init__(self, order=2):
        self.order = order  # Markov chain order
        self.markov_chain = defaultdict(list)
        
    def train(self, text):
        words = self._tokenize(text)
        for i in range(len(words) - self.order):
            key = tuple(words[i:i+self.order])
            next_word = words[i+self.order]
            self.markov_chain[key].append(next_word)
    
    def _tokenize(self, text):
        text = text.lower()
        return re.findall(r'\b\w+\b', text)
    
    def generate_sentence(self, seed=None, length=15):
        if not seed:
            seed = random.choice(list(self.markov_chain.keys()))
        current_words = list(seed)
        output = list(seed)
        
        for _ in range(length - len(seed)):
            current_tuple = tuple(current_words[-self.order:])
            next_word = random.choice(self.markov_chain.get(current_tuple, ['']))
            output.append(next_word)
            current_words.append(next_word)
            if not next_word:
                break
                
        return ' '.join(output)

# Example usage:
if __name__ == "__main__":
    bot = MarkovianChatBot(order=2)
    
    # Sample training text (can be any large body of text, like dialogue from a book)
    training_text = """
    Hello, how are you doing today? I hope you're having a great day. 
    What can I help you with today? How can I assist you? 
    It's a wonderful day for learning something new!
    """
    
    bot.train(training_text)
    
    # Generate a response based on some seed words or randomly
    print(bot.generate_sentence(seed=("how", "are")))
    
#Further explanation of chatbot:
#A Markovian ChatBot uses a Markov chain, which is a probabilistic model that predicts the next state (or word, in this case) based solely on the current state, with no consideration of the past states beyond the immediate one.
#Key Concepts:
#States: In the context of a chatbot, each state is a word or token in the text.
#Transition: A transition occurs when the chatbot moves from one word (state) to the next. The Markov chain models the probabilities of these transitions based on the training data.
#Markov Assumption: The next word in the sequence only depends on the current word, not on any words that came before it. This is the core assumption in a Markov process.
    
#1.Training the Model: During training, the chatbot reads through a corpus of text and builds a dictionary (or "model") of word transitions. The bot looks at each word in the text and keeps track of which words are likely to follow that word.
#Example:
#If the bot encounters the word "hello", it checks what word comes next. Let's say the next word is "how".
#The bot then stores that "hello" is followed by "how."
#This process continues for each word in the training text.
#After reading the entire training text, the bot would have a dictionary that looks something like this:
{
    "hello": ["how"],
    "how": ["are"],
    "are": ["you", "doing"],  # "are" can be followed by either "you" or "doing"
    "you": ["today"],
    "doing": ["well"],
    "well": ["the"],
    "the": ["weather"],
    ...
}

#2.Generating a Response:When the chatbot is asked to generate a response, it follows these steps:
#Seed Word: The bot starts with a seed word provided by the user (e.g., "hello").
#Word Selection: It looks up this word in the dictionary it built during training and selects a next word. If there's only one possible word (like "how" after "hello"), it will always pick that word. If there are multiple possible words, it chooses randomly from the list of possible next words.
#Building the Response: The bot repeats this process to build a sequence of words. It takes the word it just picked and uses it as the next seed word, then chooses another word to follow it, and so on.
#For example, if the bot starts with the seed word "hello", it follows the transitions like this:
#"hello" → "how"
#"how" → "are"
#"are" → "you"
#"you" → "today"
#"today" → and so on...
#This builds a sentence-like response, which is returned to the user.

#3.Randomness and Repetition:Because the next word is chosen randomly (from the list of possible words following the current word), the chatbot's responses can vary, even with the same input. However, with small training data, the bot might produce repetitive or limited responses because it doesn't have many options for word transitions.

#4. Limitations:
#Context-Free: A major limitation of Markov chains is that they don't remember previous states beyond the immediate one. This means the bot won't maintain context across longer conversations, and its responses can feel disjointed.
#Training Data Dependency: The quality and coherence of the chatbot's responses depend heavily on the training data. If the training data is very small or repetitive, the chatbot will mirror that in its responses.
#Short Memory: Since it only looks at the current word, it can't remember anything from earlier in the sentence or conversation, which makes it prone to producing repetitive loops or nonsensical responses.

```
